```python
import matplotlib.pylab as plt
import json
import os
import numpy as np


with open('../log/06_10_01_naive_RGB.json', 'r') as f:
    data = json.load(f)

ttv = ['train', 'test', 'val']

for s in ttv:

    plt.figure(figsize=(7,5))
    idx_min = np.argmin(data[s]['loss'])


    plt.plot(data[s]['loss'])
    plt.scatter(x=idx_min, y=data[s]['loss'][idx_min])


    plt.title(f'{s.upper()} Loss')
    plt.text(x=idx_min+0.1, y=data[s]['loss'][idx_min], s=f"Min:{round(data[s]['loss'][idx_min],4)}")

    plt.xlabel('Epochs')
    plt.ylabel('Loss')

    plt.savefig(f'./naive_{s}_loss.png')
    plt.show()
    
```

```python
with open('../log/06_10_01_naive_RGB.json', 'r') as f:
    data = json.load(f)

ttv = ['train', 'test', 'val']

for s in ttv:

    plt.figure(figsize=(7,5))
    idx_max = np.argmax(data[s]['acc'])

    plt.plot(data[s]['acc'])
    plt.scatter(x=idx_max, y=data[s]['acc'][idx_max])

    plt.title(f'{s.capitalize()} Accuracy')
    plt.text(x=idx_max+0.1, y=data[s]['acc'][idx_max], s=f"Max:{round(data[s]['acc'][idx_max],4)}")

    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')

    plt.savefig(f'./naive_{s}_acc.png')
    plt.show()
```
