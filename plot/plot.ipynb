```python
import matplotlib.pylab as plt
import json
import os

with open('../log/06_10_01_naive_RGB.json', 'r') as f:
    data = json.load(f)

ttv = ['train', 'test', 'val']

for s in ttv:

    plt.figure(figsize=(7,5))

    plt.plot(data[s]['loss'])
    plt.title(f'{s.upper()} Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')

    plt.savefig(f'./naive_{s}_loss.png')
    plt.show()
    
```

```python
os.makedirs('./plot', exist_ok=True)

with open('../log/06_10_01_naive_RGB.json', 'r') as f:
    data = json.load(f)

ttv = ['train', 'test', 'val']

for s in ttv:

    plt.figure(figsize=(7,5))

    plt.plot(data[s]['acc'])
    plt.title(f'{s.capitalize()} Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.savefig(f'./naive_{s}_acc.png')
    plt.show()
```
